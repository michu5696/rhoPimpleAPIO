```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apio_io_48_32M=np.loadtxt("apio_io_48_32M.dat")
orig_io_48_32M=np.loadtxt("orig_io_48_32M.dat")
apio_io_96_32M=np.loadtxt("apio_io_96_32M.dat")
orig_io_96_32M=np.loadtxt("orig_io_96_32M.dat")
apio_io_96_64M=np.loadtxt("apio_io_96_64M.dat")
orig_io_96_64M=np.loadtxt("orig_io_96_64M.dat")
apio_io_144_64M=np.loadtxt("apio_io_144_64M.dat")
orig_io_144_64M=np.loadtxt("orig_io_144_64M.dat")
```

```python
fig,ax=plt.subplots(1,2,figsize=(8,2))
ax[0].plot(apio_io_48_32M/(32000000*48),label="single-node")
ax[0].plot(apio_io_96_32M/(32000000*96),label="2 nodes")
ax[0].plot(apio_io_144_64M/(64000000*144),label="3 nodes")
ax[1].plot(orig_io_48_32M/(32000000*48),label="single-node")
ax[1].plot(orig_io_96_32M/(32000000*96),label="2 nodes")
ax[1].plot(orig_io_144_64M/(64000000*144),label="3 nodes")
ax[1].legend()
ax[0].legend()
```
