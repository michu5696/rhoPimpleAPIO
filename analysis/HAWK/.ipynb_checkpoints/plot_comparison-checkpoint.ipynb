```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apio_io_48_32M=np.loadtxt("apio_io_256_32M.dat")
orig_io_48_32M=np.loadtxt("orig_io_512_32M.dat")
apio_io_96_32M=np.loadtxt("apio_io_1024_32M.dat")
orig_io_96_32M=np.loadtxt("orig_io_512_64M.dat")
apio_io_96_64M=np.loadtxt("apio_io_1024_64M.dat")
orig_io_96_64M=np.loadtxt("orig_io_2048_64M.dat")
apio_io_144_64M=np.loadtxt("apio_io_1024_128M.dat")
orig_io_144_64M=np.loadtxt("orig_io_2048_128M.dat")
apio_io_96_64M=np.loadtxt("apio_io_4096_128M.dat")
orig_io_96_64M=np.loadtxt("orig_io_2048_256M.dat")
apio_io_9=np.loadtxt("apio_io_4096_256M.dat")
orig_io_8192_256M=np.loadtxt("orig_io_8192_256M.dat")
```

```python
fig,ax=plt.subplots(1,2,figsize=(8,2))
ax[0].plot(apio_io_48_32M/(32000000*48),label="48 procs")
ax[0].plot(apio_io_96_32M/(32000000*96),label="96 procs")
ax[0].plot(apio_io_144_64M/(64000000*144),label="144 procs")
ax[1].plot(orig_io_48_32M/(32000000*48),label="48 procs")
ax[1].plot(orig_io_96_32M/(32000000*96),label="96 procs")
ax[1].plot(orig_io_144_64M/(64000000*144),label="144 procs")
ax[1].legend()
ax[0].legend()
```
