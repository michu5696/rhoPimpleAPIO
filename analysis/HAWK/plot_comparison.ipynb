```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
apio_io_256_32M=np.loadtxt("apio_io_256_32M.dat")
orig_io_256_32M=np.loadtxt("orig_io_256_32M.dat")

orig_io_512_32M=np.loadtxt("orig_io_512_32M.dat")
apio_io_512_32M=np.loadtxt("apio_io_512_32M.dat")

apio_io_1024_32M=np.loadtxt("apio_io_1024_32M.dat")
orig_io_1024_32M=np.loadtxt("orig_io_1024_32M.dat")

orig_io_512_64M=np.loadtxt("orig_io_512_64M.dat")
apio_io_512_64M=np.loadtxt("apio_io_512_64M.dat")

apio_io_1024_64M=np.loadtxt("apio_io_1024_64M.dat")
orig_io_1024_64M=np.loadtxt("orig_io_1024_64M.dat")

orig_io_2048_64M=np.loadtxt("orig_io_2048_64M.dat")
apio_io_2048_64M=np.loadtxt("apio_io_2048_64M.dat")

apio_io_1024_128M=np.loadtxt("apio_io_1024_128M.dat")
orig_io_1024_128M=np.loadtxt("orig_io_1024_128M.dat")

orig_io_2048_128M=np.loadtxt("orig_io_2048_128M.dat")
apio_io_2048_128M=np.loadtxt("apio_io_2048_128M.dat")

apio_io_2048_256M=np.loadtxt("apio_io_2048_256M.dat")
orig_io_2048_256M=np.loadtxt("orig_io_2048_256M.dat")

apio_io_4096_256M=np.loadtxt("apio_io_4096_256M.dat")
orig_io_4096_256M=np.loadtxt("orig_io_4096_256M.dat")
```

```python
fig,ax=plt.subplots(2,1,figsize=(14,8))
ax[0].plot(apio_io_256_32M/(32000000*256),label="256 procs")
ax[0].plot(apio_io_512_32M/(32000000*512),label="512 procs")
ax[0].plot(apio_io_1024_32M/(32000000*1024),label="1024 procs")
ax[0].plot(apio_io_512_64M/(64000000*512),c='tab:orange')
ax[0].plot(apio_io_1024_64M/(64000000*1024),c='tab:green')
ax[0].plot(apio_io_2048_64M/(64000000*2048),label="2048 procs")

ax[1].plot(orig_io_256_32M/(32000000*256),label="256 procs")
ax[1].plot(orig_io_512_32M/(32000000*512),label="512 procs")
ax[1].plot(orig_io_1024_32M/(32000000*1024),label="1024 procs")
ax[1].plot(orig_io_2048_64M/(64000000*2048),label="2048 procs")

ax[1].legend()
ax[0].legend()
```
